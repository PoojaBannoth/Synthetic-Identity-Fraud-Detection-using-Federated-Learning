```python
# import module
import random as r
  
def phoneGen():
    ph_no = ""
      
    # the first number should be in the range of 6 to 9
    ph_no += str((r.randint(6, 9)))
      
    # the for loop is used to append the other 9 numbers.
    # the other 9 numbers can be in the range of 0 to 9.
    for i in range(1, 10):
        ph_no += str((r.randint(0, 9)))
      
    return ph_no
ph_nos = []
for i in range(50000):
    ph_nos.append(phoneGen())


df2 = pd.DataFrame(ph_nos)
df2.rename( columns={0:'Mobile'}, inplace=True )
df2
```

```python
org_data = pd.concat([df, df2], axis = 1)

org_data.head(100)
```

```python
org_data = org_data[['name', 'birthdate', 'Mobile', 'mail',  'address']]
org_data.head(100)
```

```python
org_data.rename( columns={'name':'Name', 'birthdate': 'DoB', 'mail': 'Email', 'address': 'Location'}, inplace=True)
org_data.head(5)
```

```python
print(len(org_data.index))
org_data = org_data.drop_duplicates()
print(len(org_data.index))
```

```python
f1 = org_data["Name"]
f2 = org_data["DoB"]
f3 = org_data["Email"]
f4 = org_data["Mobile"]
f5 = org_data["Location"]

f1 = f1.sample(frac =0.2, random_state=1).reset_index()
f2 = f2.sample(frac =0.2, random_state=2).reset_index()
f3 = f3.sample(frac =0.2, random_state=3).reset_index()
f4 = f4.sample(frac =0.2, random_state=4).reset_index()
f5 = f5.sample(frac =0.2, random_state=5).reset_index()


syn = pd.concat([f1, f2], axis = 1)
syn = pd.concat([syn, f3], axis = 1)
syn = pd.concat([syn, f4], axis = 1)
syn = pd.concat([syn, f5], axis = 1)

syn = syn[['Name', 'DoB', 'Mobile', 'Email', 'Location']]
print("done2")
print(len(syn.index))
print(syn.head(5))
```

```python
syn_data = pd.concat([org_data, syn], axis = 0, ignore_index=True)
print(len(syn_data.index))

syn_data = syn_data.drop_duplicates()
print(len(syn_data.index))
```

```python
import re
# Remove punctuation and apply case folding
def preprocessor(text):
    text = re.sub('<[^>]*>', '', str(text).lower())
    emoticons = re.findall('\n', text)
    text = (re.sub('[\W]+', ' ', text.lower()) + ' ' .join(emoticons).replace(',', ''))

    return text
```

```python
syn_data['Location'] = syn_data['Location'].apply(preprocessor)

syn_data.head(100)
```

```python
syn_data['label'] = 0
for ind in syn_data.index:
    if ind >= len(org_data.index):
        syn_data.loc[ind, 'label'] = 1
```
